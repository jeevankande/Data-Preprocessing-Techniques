```python
import pandas as pd 
import numpy as np
```

```python
df_claimants = pd.read_csv("H:\study\claimants.csv")
```

```python
df_claimants.isna().sum() #to get na val from each col
```

```python
from sklearn.impute import SimpleImputer

#mean Imputr for CLMAGE, CLMSEX
#mode Imputer for CLMINSUR, SEATBELT
```

```python
mean_imputer = SimpleImputer(missing_values=np.nan, strategy= 'mean')

median_imputer = SimpleImputer(missing_values= np.nan, strategy='median')
```

```python
df_claimants[['CLMAGE', 'CLMSEX']] = pd.DataFrame(mean_imputer.fit_transform
                                                  (df_claimants[['CLMAGE', 'CLMSEX']])) 
```

```python
df_claimants[['CLMINSUR','SEATBELT']] = pd.DataFrame(median_imputer.fit_transform
                                                     (df_claimants[['CLMINSUR', 'SEATBELT']]))
```
