```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() #standedised scaler
```

```python
datset = np.array(seeds_df.Area).reshape(-1,1)

scaler.fit(datset)
```

```python
#standardized_Area = scaler.Transform(datset)
```

```python
standardized_Area = scaler.transform(datset)
```

```python
seeds_df['standardized_Area'] = scaler.transform(datset)
```

```python
seeds_df
```

```python
stats.probplot(seeds_df.standardized_Area, dist = 'norm', plot = plt)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
N_scalar = MinMaxScaler()  #normalised scalar
```

```python
data = np.array(seeds_df.Area).reshape(-1,1)
```

```python
N_scalar.fit(data)
```

```python
seeds_df['normalized_dataset'] = N_scalar.transform(data)
```

```python
seeds_df
```

```python
stats.probplot(seeds_df.normalized_dataset, dist = 'norm', plot = plt)
plt.show()
```
